```python
# import fastas from Benchmark Genes folder
import os
import sys
import Bio
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

# get the path to the Benchmark Genes folder
path = "/Users/rishabjain/Desktop/Research/choformer/benchmarking/Benchmark Genes"

# get the list of all the files in the Benchmark Genes folder
files = os.listdir(path)


# get the list of all the fasta files in the Benchmark Genes folder
fastas = [file for file in files if file.endswith(".fasta")]

# get all the genes from each file (some files can have multiple)

genes = []
for fasta in fastas:
    with open(path + "/" + fasta, "r") as handle:
        for record in SeqIO.parse(handle, "fasta"):
            genes.append(record)

# get the list of all the genes
gene_list = [gene.id for gene in genes]

# get the list of all the gene sequences
gene_sequences = [str(gene.seq) for gene in genes]

# get the list of all the gene lengths
gene_lengths = [len(gene) for gene in gene_sequences]

# get the list of all the gene descriptions
gene_descriptions = [gene.description for gene in genes]

# write a final fasta file with all the genes
with open("all_genes.fasta", "w") as handle:
    for gene in genes:
        SeqIO.write(gene, handle, "fasta")

# write a final text file with all the gene descriptions
with open("all_gene_descriptions.txt", "w") as handle:
    for gene in gene_descriptions:
        handle.write(gene + "\n")

```

```python
import numpy as np
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO

# Codon usage table
frequency = {
    "A": (["GCG", "GCA", "GCT", "GCC"], [0.11, 0.23, 0.26, 0.40]),
    "R": (["AGG", "AGA", "CGG", "CGA", "CGT", "CGC"], [0.20, 0.20, 0.21, 0.11, 0.08, 0.19]),
    "N": (["AAT", "AAC"], [0.46, 0.54]),
    "D": (["GAT", "GAC"], [0.46, 0.54]),
    "C": (["TGT", "TGC"], [0.45, 0.55]),
    "*": (["TGA", "TAG", "TAA"], [0.52, 0.20, 0.28]),
    "Q": (["CAG", "CAA"], [0.75, 0.25]),
    "E": (["GAG", "GAA"], [0.58, 0.42]),
    "G": (["GGG", "GGA", "GGT", "GGC"], [0.25, 0.25, 0.16, 0.34]),
    "H": (["CAT", "CAC"], [0.41, 0.59]),
    "I": (["ATA", "ATT", "ATC"], [0.16, 0.36, 0.48]),
    "L": (["TTG", "TTA", "CTG", "CTA", "CTT", "CTC"], [0.13, 0.07, 0.41, 0.07, 0.13, 0.20]),
    "K": (["AAG", "AAA"], [0.58, 0.42]),
    "M": (["ATG"], [1.0]),
    "F": (["TTT", "TTC"], [0.45, 0.55]),
    "P": (["CCG", "CCA", "CCT", "CCC"], [0.11, 0.27, 0.28, 0.33]),
    "S": (["AGT", "AGC", "TCG", "TCA", "TCT", "TCC"], [0.15, 0.24, 0.06, 0.15, 0.18, 0.22]),
    "T": (["ACG", "ACA", "ACT", "ACC"], [0.12, 0.28, 0.24, 0.36]),
    "W": (["TGG"], [1.0]),
    "Y": (["TAT", "TAC"], [0.43, 0.57]),
    "V": (["GTG", "GTA", "GTT", "GTC"], [0.47, 0.11, 0.18, 0.24])
}

# Reverse translate protein sequence to DNA sequence
def reverse_translate(protein_sequence):
    dna_sequence = ""
    for amino_acid in protein_sequence:
        codons, probs = frequency[amino_acid]
        # Normalize the probabilities to sum to 1
        probs = np.array(probs) / np.sum(probs)
        dna_sequence += codons[np.random.choice(len(codons), p=probs)]
    return dna_sequence

# Read the proteins from all_genes.fasta
proteins = []
with open("all_genes.fasta", "r") as handle:
    for record in SeqIO.parse(handle, "fasta"):
        proteins.append(record)

# Reverse translate each gene sequence
gene_sequences_dna = [reverse_translate(gene) for gene in proteins]

# Write a final FASTA file with all the genes
with open("all_genes_dna.fasta", "w") as handle:
    for i in range(len(gene_sequences_dna)):
        gene = SeqRecord(Seq(gene_sequences_dna[i]), id=f"gene_{i+1}", description=f"gene_{i+1}")
        SeqIO.write(gene, handle, "fasta")
```
