```python
from Bio import SeqIO

# Function to count sequences in a FASTA file
def count_sequences(fasta_file):
    count = 0
    for ix, _ in enumerate(SeqIO.parse(fasta_file, "fasta")):
        count += 1
        df.loc[ix, "protein"] = str(_.seq)
        # print(_.seq)
        # break
    return count

# Example usage
fasta_file = "your_protein_sequences.fasta"  # Replace with your FASTA file path
sequence_count = count_sequences("cleaned_protein_sequences.fasta")

print(f"The number of sequences in the FASTA file is: {sequence_count}")
```
